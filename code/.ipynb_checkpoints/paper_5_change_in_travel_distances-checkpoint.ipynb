## Read Data

### Main data

```python
cbg_connections = pd.read_csv("../data/cbg_connections_processed.csv", parse_dates=['date'],
                              dtype={'visitor_home_cbg': str, 'GEOID_origin': str, 'GEOID_dest': str,
                                     'visitor_home_county': str, 'visitor_dest_county': str})
```

```python
# select nyc local visitors only
cbg_connections = cbg_connections[cbg_connections['visitor_home_county'].isin(['36005','36047','36061','36081','36085'])]
```

```python
# cbg_connections.info()
```

### Income data

```python
# nyc income data by CBG
nyc_acs_income_data = pd.read_csv("../data/income/nyc_acs_income_data_2019.csv", dtype={'cbg_id':str, 'county_id':str})
```

```python
nyc_acs_income_data.info()
```

```python
# income of the parks' neighborhoods (processed from ArcGIS Pro)
nyc_parks_nbhd_income = pd.read_csv("../data/income/nyc_parks_nbhd_income.csv")
```

```python
nyc_parks_nbhd_income.info()
```

## Change in Travel Distance

```python
# park CBG income
cbg_connections_income = pd.merge(cbg_connections, nyc_parks_nbhd_income, on='safegraph_place_id', how='left')

# visitor CBG income
cbg_connections_income = pd.merge(cbg_connections_income, nyc_acs_income_data, left_on='GEOID_origin', right_on='cbg_id',
                                  how='left', suffixes=['_dest','_origin'])
```

```python
# the major park types that attracted more than 90% of the total visits
major_park_types = ['Community Park', 'Flagship Park', 'Jointly Operated Playground', 'Nature Area',
                    'Neighborhood Park', 'Playground', 'Recreation Field/Courts', 'Triangle/Plaza']

#select the data for major park types and for the years of 2019 and 2020
cbg_connections_income_major = cbg_connections_income[cbg_connections_income['park_type'].isin(major_park_types)
                                                      & cbg_connections_income['year'].isin([2019,2020])]
```

```python
# expand data to make each record as an individual
cbg_connections_income_major = cbg_connections_income_major.loc[cbg_connections_income_major.index.repeat(cbg_connections_income_major['visitor_count'])].reset_index()
```

## Change in travel distance by park type

### Summary Table

```python
data = cbg_connections_income_major
topic_col = 'park_type'
value_col = 'distance_geod'
by_income_group = False

summary_distance = pvcr.summarize_travel_distance(data, topic_col, value_col, by_income_group)
```

```python
# calculate the mean value of the confidence interval
summary_distance["mean_diff_pct_cim"] = (summary_distance["diff_ub_pct"] - summary_distance["diff_lb_pct"]) / 2
```
