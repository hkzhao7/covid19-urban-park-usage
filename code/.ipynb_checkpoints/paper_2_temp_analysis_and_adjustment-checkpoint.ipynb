## Fit a model for park visits vs temp

```python
park_visits_monthly_total = park_visits_total_monthly.groupby(['date', 'year', 'month'], as_index=False)['visitor_count'].sum()
```

```python
park_visits_temp = pd.merge(park_visits_monthly_total, nyc_avg_temp_monthly,
                            left_on='date', right_on='DATE')
park_visits_temp = park_visits_temp[park_visits_temp['date'] < datetime.datetime(2020,3,1)]
```

```python
park_visits_temp.head()
```

### Least squares curve fit

```python
X = park_visits_temp.sort_values(['TAVG'])['TAVG'].to_numpy()
y = park_visits_temp.sort_values(['TAVG'])['visitor_count'].to_numpy()
```

```python
popt_1, pcov_1, r2_1 = temp_adj.fit_curve(temp_adj.func_linear, X, y, print_label=" (Linear): ")
popt_2, pcov_2, r2_2 = temp_adj.fit_curve(temp_adj.func_exp, X, y, print_label=" (Exp): ")
popt_3, pcov_3, r2_3 = temp_adj.fit_curve(temp_adj.func_poly2, X, y, print_label=" (2-Degree Poly): ")
popt_4, pcov_4, r2_4 = temp_adj.fit_curve(temp_adj.func_poly3, X, y, print_label=" (3-Degree Poly): ")
```

```python
# temp_adj.plot_least_square_fit_result(X, y, popt_1, popt_2, popt_3, popt_4)
```

### Gaussian process fit

```python
# training data
X = park_visits_temp.sort_values(['TAVG'])['TAVG'].to_numpy().reshape(-1,1)
y = park_visits_temp.sort_values(['TAVG'])['visitor_count'].to_numpy().ravel()
```

```python
x, y_pred, sigma, pipe = temp_adj.gaussian_process_regression(X, y)
```

```python
r2_gp = pipe.score(X, y)
```

```python
r2_gp
```

### Combined Regression Results Plot

```python
temp_adj.plot_all_model_results(X, y, x, y_pred, sigma, r2_gp,
                                popt_1, popt_2, popt_3, popt_4,
                                r2_1, r2_2, r2_3, r2_4)
```
